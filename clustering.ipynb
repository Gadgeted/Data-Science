```python
import pandas
# we read bank data 
data = pandas.read_csv("datasets/Mall_Customers.csv")
# above data variable has the our data
data
# eda - exploratory data analysis
```

```python
data.dropna(inplace=True)
array = data.values
x = array[:, 2:5] # 
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=10, random_state=42)  #  
model.fit(x)
```

```python
means = model.cluster_centers_
clusters = pandas.DataFrame(means, columns=['Age', 'income', 'score'])
clusters
```

```python
data['label'] = model.labels_
data
```

```python
cluster2 = data[data['label'] == 6]
cluster2
```
